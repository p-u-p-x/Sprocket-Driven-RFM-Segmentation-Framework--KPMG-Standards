```python
# Clean RFM_VALUE column (handle any non-numeric values if they exist)
df_rfm_value['RFM_VALUE'] = pd.to_numeric(df_rfm_value['RFM_VALUE'], errors='coerce')

# Define customer title mapping based on RFM_VALUE
def assign_customer_title(rfm_value):
    if pd.isna(rfm_value):
        return 'Unknown'
    rfm_value = int(rfm_value)
    if rfm_value == 444:
        return 'Platinum'
    elif 433 <= rfm_value <= 444:
        return 'Very Loyal'
    elif 421 <= rfm_value < 433:
        return 'Becoming Loyal'
    elif 344 <= rfm_value < 421:
        return 'Recent'
    elif 212 <= rfm_value < 344:
        return 'High Risk'
    elif 124 <= rfm_value < 212:
        return 'Almost Lost'
    elif 112 <= rfm_value < 124:
        return 'Evasive'
    elif rfm_value == 111:
        return 'Lost'
    else:
        return 'Other'

# Apply the new customer title classification
df_rfm_value['New_Customer_Title'] = df_rfm_value['RFM_VALUE'].apply(assign_customer_title)

# Validation - Compare old and new titles
print("\nValidation of Customer Title Assignment:")
validation_df = df_rfm_value[['RFM_VALUE', 'CustomerTitle', 'New_Customer_Title']].head(10)
display(validation_df)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Merge RFM data with transactions to get wealth segments
merged_df = pd.merge(
    df_rfm_value,
    df_transactions[['customer_id', 'wealth_segment', 'Profit']],
    left_on='Row Labels',
    right_on='customer_id',
    how='left'
)

# Clean data - fill missing wealth segments
merged_df['wealth_segment'] = merged_df['wealth_segment'].fillna('Unknown')

# Create pivot table
pivot = pd.pivot_table(
    merged_df,
    values='Profit',
    index='New_Customer_Title',
    columns='wealth_segment',
    aggfunc='mean',
    fill_value=0
)

# Ensure all customer titles are included
all_titles = [
    'Platinum', 'Very Loyal', 'Becoming Loyal', 'Recent',
    'High Risk', 'Almost Lost', 'Evasive', 'Lost', 'Unclassified'
]
pivot = pivot.reindex(all_titles, fill_value=0)

# Get existing wealth segments and order them by value
existing_wealth_segments = [seg for seg in ['High Net Worth', 'Affluent Customer', 'Mass Customer', 'Unknown']
                          if seg in pivot.columns]
pivot = pivot[existing_wealth_segments]

# Professional blue-gray palette (NO ORANGE)
wealth_palette = {
    'High Net Worth': '#2C4A6B',  # Dark navy
    'Affluent Customer': '#4B7B9D',  # Steel blue
    'Mass Customer': '#7BA5C8',  # Light slate blue
    'Unknown': '#B0B8B4'  # Cool gray
}
custom_colors = [wealth_palette[seg] for seg in existing_wealth_segments]

# Create figure with two subplots
plt.style.use('default')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8), gridspec_kw={'width_ratios': [4, 5]})

# Formatting parameters
title_params = {'pad': 15, 'fontsize': 14, 'fontweight': 'semibold', 'color': '#2F2F2F'}
label_params = {'labelpad': 10, 'fontsize': 11, 'color': '#2F2F2F'}

# First chart - High Value Customers
pivot.iloc[:4].plot(kind='bar', stacked=True, ax=ax1, color=custom_colors,
                   edgecolor='white', linewidth=0.7, width=0.75)
ax1.set_title('Premium Customer Segments', **title_params)
ax1.set_ylabel('Average Profit ($)', **label_params)
ax1.set_xlabel('', **label_params)
ax1.yaxis.set_major_formatter('${x:,.0f}')
ax1.grid(axis='y', linestyle=':', alpha=0.4, color='#CCCCCC')

# Second chart - At-Risk Customers
pivot.iloc[4:].plot(kind='bar', stacked=True, ax=ax2, color=custom_colors,
                   edgecolor='white', linewidth=0.7, width=0.75)
ax2.set_title('At-Risk Customer Segments', **title_params)
ax2.set_xlabel('', **label_params)
ax2.yaxis.set_major_formatter('${x:,.0f}')
ax2.grid(axis='y', linestyle=':', alpha=0.4, color='#CCCCCC')

# Format both axes
for ax in [ax1, ax2]:
    ax.tick_params(axis='x', rotation=45, labelsize=10, colors='#2F2F2F')
    ax.tick_params(axis='y', labelsize=10, colors='#2F2F2F')
    for spine in ['top', 'right', 'left']:
        ax.spines[spine].set_visible(False)
    ax.spines['bottom'].set_color('#DDDDDD')
    ax.set_axisbelow(True)

# Add value labels
for ax in [ax1, ax2]:
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(
                f'${height:,.0f}',
                (p.get_x() + p.get_width() / 2., p.get_y() + height / 2),
                ha='center',
                va='center',
                fontsize=9,
                color='white' if height > max(pivot.max())/2 else '#2F2F2F',
                fontweight='medium'
            )

# Add unified legend
handles, labels = ax1.get_legend_handles_labels()
legend = fig.legend(handles, labels, title='Wealth Segment',
          bbox_to_anchor=(0.5, 1.02), loc='lower center',
          ncol=4, frameon=True, facecolor='white',
          edgecolor='#DDDDDD', title_fontproperties={'weight':'semibold', 'size':11},
          fontsize=10, borderpad=1)

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(top=0.85, wspace=0.2, bottom=0.15)

# Display styled pivot table with high contrast
print("Average Profit by Customer RFM Classification and Wealth Segment:")

def color_column(col):
    return [f'background-color: {wealth_palette[col.name]}; color: white' for _ in col]

display(
    pivot.style
    .format("${:,.2f}")
    .apply(color_column)
    .set_properties(**{
        'border': '1px solid #DDDDDD',
        'font-size': '11px',
        'text-align': 'center'
    })
    .set_table_styles([{
        'selector': 'th',
        'props': [
            ('background-color', '#2C4A6B'),
            ('color', 'white'),
            ('font-weight', 'bold'),
            ('font-size', '12px'),
            ('border', '1px solid #2C4A6B')
        ]
    }, {
        'selector': 'td',
        'props': [
            ('border', '1px solid #DDDDDD')
        ]
    }])
)

plt.show()
```

```python
# Prepare and merge the data
df_transactions['customer_id'] = df_transactions['customer_id'].astype(str).str.strip()
df_rfm_value['Row Labels'] = df_rfm_value['Row Labels'].astype(str).str.strip()

merged_df = pd.merge(
    df_transactions,
    df_rfm_value[['Row Labels', 'New_Customer_Title']].rename(columns={'Row Labels': 'customer_id'}),
    on='customer_id',
    how='left'
)

merged_df['New_Customer_Title'] = merged_df['New_Customer_Title'].fillna('Unclassified')
merged_df['month'] = merged_df['transaction_date'].dt.strftime('%b')

# Pivot table prep
all_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
all_titles = ['Platinum', 'Very Loyal', 'Becoming Loyal', 'Recent',
              'High Risk', 'Almost Lost', 'Evasive', 'Lost', 'Unclassified']

pivot = (merged_df.groupby(['month', 'New_Customer_Title'])['Profit']
         .mean()
         .unstack()
         .reindex(index=all_months, columns=all_titles)
         .fillna(0))

# Display styled table
print("Average Profit by Month and Customer RFM Segment")
display(
    pivot.style
    .format("${:,.2f}")
    .background_gradient(axis=0, cmap='Blues')
    .set_properties(**{
        'font-size': '11pt',
        'text-align': 'center',
        'border': '1px solid #dddddd'
    })
    .set_table_styles([{
        'selector': 'th',
        'props': [
            ('background-color', '#3A5A78'),
            ('color', 'white'),
            ('font-weight', 'bold'),
            ('border', '1px solid #3A5A78')
        ]
    }])
)
```

```python
# Professional color palette
chart_palette = [
    '#2C4A6B', '#3E6B8E', '#5A8CB3', '#7BA5C8',
    '#9DBDDD', '#B8D0E5', '#D1E0ED', '#E0E8F0', '#EDF2F7'
]

# Visualization setup
plt.figure(figsize=(20.5, 8.5))
ax = plt.gca()

n_months = len(pivot.index)
n_titles = len(pivot.columns)
bar_width = min(0.8, 0.8 / n_titles)
index = np.arange(n_months)

for i, (title, color) in enumerate(zip(pivot.columns, chart_palette)):
    values = pivot[title]
    bars = ax.bar(
        index + i * bar_width,
        values,
        bar_width,
        label=title,
        color=color,
        edgecolor='white',
        linewidth=0.5,
        alpha=0.9
    )

    for j, val in enumerate(values):
        if val > 0:
            ax.text(
                index[j] + i * bar_width,
                val + 0.01 * pivot.values.max(),
                f'${val:,.0f}',
                ha='center',
                va='bottom',
                fontsize=9,
                color='#2F2F2F',
                fontweight='medium'
            )

# Chart formatting
plt.title('Average Profit by Month and Customer RFM Segment',
          fontsize=15, pad=20, fontweight='semibold')
plt.xlabel('Month', fontsize=12, labelpad=12)
plt.ylabel('Average Profit ($)', fontsize=12, labelpad=12)
plt.xticks(index + bar_width * (n_titles - 1) / 2, pivot.index,
           rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)
ax.yaxis.set_major_formatter('${x:,.0f}')

# Grid and borders
ax.grid(axis='y', linestyle=':', alpha=0.5)
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

# Legend
plt.legend(
    title='Customer Segment',
    loc='upper left',
    bbox_to_anchor=(1.02, 1),
    frameon=True,
    edgecolor='#dddddd'
)

plt.tight_layout()
plt.show()
```
